```python
import pandas as pd
```

```python
filename = "/content/sample_data/RailsToTrails_National Count Data_week.xlsx"
bike_df = pd.read_excel(filename)
```

```python
bike_df.head()
```

```python
bike_df.info()
```

Eliminar Columnas innecesarias

```python
bike_df = bike_df.loc[: , "Timeframe":"Unnamed: 6"]
bike_df.head()
```

Renombrar columnas

```python
bike_df=bike_df.rename(columns={"Unnamed: 6": "percent_change."},errors="raise")
bike_df.head()
```

```python
bike_df.head()
```

Valores faltantes

```python
bike_df.isna().sum()
```

Reemplazar valores faltante

```python
bike_df ["2021 Counts"]=bike_df["2021 Counts"].fillna(0)
bike_df.isna().sum()
```
