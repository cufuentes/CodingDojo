```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/mGalarnyk/Tutorial_Data/master/Pandas/CalIt2.data"
```

```python
df = pd.read_csv(url,header=None)
df.head(20)
```

```python
df = df.rename({0:"Flow_ID",1:"Date",2:"Time",3:"Count"},axis="columns")
df.head(10)
```

```python
top_10=df.sort_values(by="Count",ascending=False).head(10)
top_10
```

duda aqui:

Aplicar funcin
Para las 10 filas que aparecen arriba, utilice la funcin Apply de Pandas para restar el valor ms bajo de los 10 de todos ellos y luego obtengan el valor promedio de los recuentos resultantes


```python
top_10["Count"] = top_10["Count"] - top_10["Count"].min()
top_10
```

Otra forma: usando apply

```python
top_10_apply=df.sort_values(by="Count",ascending=False).head(10)
menor = top_10_apply["Count"].min()

top_10_apply["Conteo Ajustado"] = top_10_apply["Count"].apply(lambda x: x - menor)
top_10_apply["Conteo Ajustado"].mean()
```
